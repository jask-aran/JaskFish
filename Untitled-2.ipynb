```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Given data
return_debt = 0.08
return_equity = 0.13
SD_debt = 0.12
SD_equity = 0.2
correlation = 0.3


# Weight range
weights = np.linspace(0, 1, 500)

# Calculate portfolio return and standard deviation
portfolio_return = weights * return_equity + (1 - weights) * return_debt
portfolio_variance = (weights**2 * SD_equity**2 + 
                     (1 - weights)**2 * SD_debt**2 + 
                     2 * weights * (1 - weights) * correlation * SD_equity * SD_debt)
portfolio_std_dev = np.sqrt(portfolio_variance)



df = pd.DataFrame({
    'Weight_Debt': 1 - weights,
    'Weight_Equity': weights,
    'Portfolio_Return': portfolio_return,
    'Portfolio_Std_Dev': portfolio_std_dev
})

df.iloc[np.linspace(0, len(df) - 1, 11, dtype=int)]
```

```python
# Calculate the minimum portfolio standard deviation
min_std_dev_index = np.argmin(portfolio_std_dev)
min_std_dev = portfolio_std_dev[min_std_dev_index]
min_return = portfolio_return[min_std_dev_index]

# Plot the curve with the efficient frontier and the point where it begins
plt.figure(figsize=(7.5, 4.5))
plt.plot(portfolio_std_dev, portfolio_return, label='Opportunity Set of Debt and Equity')
plt.plot(portfolio_std_dev[min_std_dev_index:], portfolio_return[min_std_dev_index:], color='red', linestyle=(0, (5, 5)), label='Efficient Frontier')

plt.scatter(portfolio_std_dev[min_std_dev_index], portfolio_return[min_std_dev_index], 
            color='blue', zorder=5, marker='x',
            label=f'Start of Efficient Frontier\n(w_D={1-weights[min_std_dev_index]:.3f}, w_E={weights[min_std_dev_index]:.3f})')
plt.scatter(portfolio_std_dev[0], portfolio_return[0], color='green', marker='x', zorder=5, label='100% Debt')
plt.scatter(portfolio_std_dev[-1], portfolio_return[-1], color='purple', marker='x', zorder=5, label='100% Equity')

plt.xlabel('Portfolio Standard Deviation')
plt.ylabel('Portfolio Expected Return')
plt.title('Efficient Frontier of Portfolios Composed of Debt and Equity')
plt.legend(borderpad=1, loc='lower right')
plt.grid(True)
plt.show()
```

```python
risk_free_rate = 0.05

# Calculate optimal weights
cov_DE = correlation * SD_debt * SD_equity
numerator = (return_debt - risk_free_rate) * SD_equity**2 - (return_equity - risk_free_rate) * cov_DE
denominator = (return_debt - risk_free_rate) * SD_equity**2 + (return_equity - risk_free_rate) * SD_debt**2 - (return_debt + return_equity - 2 * risk_free_rate) * cov_DE
weight_debt_optimal = numerator / denominator
weight_equity_optimal = 1 - weight_debt_optimal

# Calculate expected return and standard deviation of the optimal portfolio
return_optimal = weight_debt_optimal * return_debt + weight_equity_optimal * return_equity
variance_optimal = (weight_debt_optimal**2 * SD_debt**2 + 
                   weight_equity_optimal**2 * SD_equity**2 + 
                   2 * weight_debt_optimal * weight_equity_optimal * cov_DE)
std_dev_optimal = np.sqrt(variance_optimal)

# Capital Allocation Line (CAL) extended to intersect the left edge of the plot
cal_x = np.linspace(0, 0.4, 100)
cal_y = risk_free_rate + (return_optimal - risk_free_rate) / std_dev_optimal * cal_x

plt.figure(figsize=(7.5, 4.5))
plt.plot(portfolio_std_dev, portfolio_return, label='Opportunity Set of Portfolios')
plt.plot(portfolio_std_dev[min_std_dev_index:], portfolio_return[min_std_dev_index:], color='red', linestyle=(0, (5, 5)), label='Efficient Frontier')

plt.scatter(portfolio_std_dev[min_std_dev_index], portfolio_return[min_std_dev_index], 
            color='blue', zorder=5, marker='x',
            label=f'Start of Efficient Frontier\n(w_D={1-weights[min_std_dev_index]:.3f}, w_E={weights[min_std_dev_index]:.3f})')
plt.scatter(portfolio_std_dev[0], portfolio_return[0], color='green', marker='x', zorder=5, label='100% Debt')
plt.scatter(portfolio_std_dev[-1], portfolio_return[-1], color='purple', marker='x', zorder=5, label='100% Equity')
plt.scatter(std_dev_optimal, return_optimal, color='orange', zorder=5, marker='x', 
            label=f'Optimal Portfolio \n(w_D={weight_debt_optimal:.2f}, w_E={weight_equity_optimal:.2f} )\n E(r_P)={return_optimal:.3f}, SD(P)={std_dev_optimal:.3f}')
plt.scatter(0, risk_free_rate, color='black', marker='o', zorder=5, label=f'Risk-Free Rate (r_f={risk_free_rate})')
plt.annotate('Risk-Free Rate', xy=(0, risk_free_rate), xytext=(0.02, risk_free_rate + 0.005),
             arrowprops=dict(facecolor='black', shrink=0.05))

# plt.xlim(min_std_dev - 0.02, SD_equity +0.02)  # Keeping the current plot view
# plt.ylim(return_debt - 0.01, return_equity + 0.01)  # Adjusting the y-axis for better visibility

plt.xlabel('Portfolio Standard Deviation')
plt.ylabel('Portfolio Expected Return')
plt.title('Optimal Risky Portfolio; Tangency Portfolio')

plt.grid(True)

plt.autoscale(False)
plt.plot(cal_x, cal_y, color='orange', linestyle='-', label='Capital Allocation Line (CAL)', scaley=False)
plt.legend(frameon=True)
plt.show()
```
